## Creating a key value pair 

`characters` is a dictionary which consist of key as index and value as characters. 
<br /> <br />
`index` has key as characters and value as index.

we will see why we do that in further code.

```python
all_chars = list(dict.fromkeys("Hello world")) 
characters = {} 
index = {} 

for i in range(0, len(all_chars)):
    characters[i] = all_chars[i]
    index[all_chars[i]] = i
print(characters)
print(index)

def encode(string):
    encoded = []
    input_list = list(string)
    for i in range(0, len(input_list)):
        current_input = input_list[i]
        encoded.append(
            index[current_input]
        )
    return encoded

def decoder(string):
    decoded = []
    for i in range(0, len(string)):
        current_pos = string[i]
        decoded.append(
            characters[current_pos]
        )
    return "".join(decoded)

print(encode("Hello"))
print(decoder([0, 1, 2, 2, 3]))
```
